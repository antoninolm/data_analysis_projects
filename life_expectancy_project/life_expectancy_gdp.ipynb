This project explores the relationship between life expectancy and GDP across six countries (Chile, China, Germany, Mexico, USA, Zimbabwe) from 2000‚Äì2015.
We clean the dataset, run descriptive summaries, check correlations, and visualize key trends.

# Libraries & Data Load

We start by importing core libraries and loading the CSV file.

```python
# Core libraries for data manipulation
import pandas as pd
import numpy as np
import csv

# Plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("all_data.csv")
```

‚úÖ At this stage, the raw dataset is loaded with original column names.

# Clean & Prepare

We rename long column names to simpler snake_case, check datatypes, and confirm there are no missing values.

```python
# ======================
# 2) Clean & prepare
# ======================
df = df.rename(columns={
    "Life expectancy at birth (years)": "life_expectancy",
    "Country": "country",
    "Year": "year",
    "GDP": "gdp"
})
df.info()
```

```python
# Quick shape & dtypes
print("Rows, Cols:", df.shape)
print(df.dtypes, "\n")

# Basic missing check
print("Missing values per column:\n", df.isna().sum(), "\n")
```

üìå Output shows:

96 rows √ó 4 columns

Correct types (int64 for year, float64 for life expectancy & GDP)

No missing values

# Quick Descriptive Insights

We summarize per-country changes over time: starting/ending life expectancy, GDP growth, and life expectancy gains.

```python
by_country = df.groupby("country").agg(
    years_min=("year", "min"),
    years_max=("year", "max"),
    le_start=("life_expectancy", "first"),
    le_end=("life_expectancy", "last"),
    le_min=("life_expectancy", "min"),
    le_max=("life_expectancy", "max"),
    gdp_start=("gdp", "first"),
    gdp_end=("gdp", "last")
).reset_index()

# Add life expectancy change
by_country["le_change"] = by_country["le_end"] - by_country["le_start"]

# GDP CAGR over period
years_span = by_country["years_max"] - by_country["years_min"]
by_country["gdp_cagr"] = ((by_country["gdp_end"] / by_country["gdp_start"]) ** (1 / years_span) - 1) * 100

print("Per-country summary:")
print(by_country[["country","le_start","le_end","le_change","gdp_cagr"]], "\n")
```

üîç Findings:

China shows the highest GDP CAGR (~15.9%/yr).

Zimbabwe improved life expectancy the most (+14.7 yrs), despite modest GDP.

Developed countries like Germany and USA show smaller but steady improvements.

# Correlation Analysis

We check whether GDP (log scale) and life expectancy are correlated.

```python
def corr_log_gdp_le(sub):
    if sub["gdp"].gt(0).all() and sub["life_expectancy"].notna().all():
        return np.corrcoef(np.log10(sub["gdp"]), sub["life_expectancy"])[0,1]
    return np.nan

corrs = df.groupby("country").apply(corr_log_gdp_le).rename("corr_log_gdp_le").reset_index()
print("Correlation log10(GDP) vs life expectancy by country:")
print(corrs, "\n")
```

üìå All countries show strong positive correlations (0.92‚Äì0.98): higher GDP tends to align with higher life expectancy.

# Visualizations

We plot multiple views of the data.

## Life Expectancy Over Time

```python
palette = sns.color_palette("tab10", df["country"].nunique())

plt.figure(figsize=(9,6))
sns.lineplot(data=df, x="year", y="life_expectancy", hue="country", palette=palette, marker="o")
plt.title("Life Expectancy Over Time (2000‚Äì2015)")
plt.xlabel("Year"); plt.ylabel("Years")
plt.legend(title="Country", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout(); plt.show()
```

üëâ All countries improve; Zimbabwe rebounds sharply after 2005.

## GDP Over Time (Log Scale)

```python
plt.figure(figsize=(9,6))
sns.lineplot(data=df, x="year", y="gdp", hue="country", palette=palette, marker="o")
plt.yscale("log")
plt.title("GDP Over Time (Log Scale)")
plt.xlabel("Year"); plt.ylabel("GDP (current US$, log scale)")
plt.legend(title="Country", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout(); plt.show()
```

üëâ Huge GDP growth in China; developed economies remain stable.

## GDP vs Life Expectancy

```python
plt.figure(figsize=(9,6))
sns.scatterplot(data=df, x="gdp", y="life_expectancy", hue="country", palette=palette, s=80, alpha=0.8)
plt.xscale("log")
plt.title("Life Expectancy vs GDP")
plt.xlabel("GDP (current US$, log scale)")
plt.ylabel("Life expectancy (years)")
plt.legend(title="Country", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout(); plt.show()
```

üëâ Nonlinear relationship: gains in life expectancy flatten at high GDP levels.

## Per-Country Mini Trends

```python
g = sns.FacetGrid(df, col="country", col_wrap=3, height=3.2, sharey=False)
g.map_dataframe(sns.lineplot, x="year", y="life_expectancy", marker="o")
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Year", "Life exp (yrs)")
plt.tight_layout(); plt.show()
```

üëâ Small multiples reveal Zimbabwe‚Äôs sharp turnaround, China‚Äôs steady rise, and stabilization in developed countries.

# Final Takeaways

We summarize the key findings per country.

```python
print("Quick takeaways:")
for _, row in by_country.iterrows():
    print(f"- {row['country']}: life expectancy {row['le_start']:.1f} ‚Üí {row['le_end']:.1f} "
          f"(Œî {row['le_change']:+.1f} yrs), GDP CAGR ‚âà {row['gdp_cagr']:.2f}%/yr over {int(row['years_max']-row['years_min'])} years.")
print("\nPositive correlations (log GDP vs life expectancy) suggest higher GDP tends to align with higher life expectancy, "
      "though magnitudes differ by country and improvements can occur even with modest GDP growth (e.g., Zimbabwe‚Äôs rebound).")
```

üìä Key Takeaways:

China: fastest GDP growth (+15.9%/yr), +4.4 yrs life expectancy

Zimbabwe: strongest life expectancy recovery (+14.7 yrs), moderate GDP growth

USA & Germany: stable GDP, modest life expectancy gains

Chile & Mexico: steady improvements, though smaller than China
