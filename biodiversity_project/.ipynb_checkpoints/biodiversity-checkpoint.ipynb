In this project, we analyze biodiversity data from US national parks.  
We combine **species metadata** with **observation counts** to study protection status, categories, and observation frequency.  

---

# Main Set Up

## Step 0 â€” Imports

```python
# ======================
# STEP 0 â€” Imports
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 0.1 â€” Load CSVs

We load the two datasets:
- `species_info.csv`: metadata about species (category, names, conservation status)
- `observations.csv`: counts of observations per species in national parks

We also treat common placeholders like `"nan"` or empty strings as missing values.

```python
# ======================
# STEP 0.1 â€” Load CSVs
#   - Treat common string placeholders as missing values
# ======================
species_raw = pd.read_csv("species_info.csv", na_values=["nan", "NaN", "None", ""])
obs_raw     = pd.read_csv("observations.csv", na_values=["nan", "NaN", "None", ""])

print("Loaded shapes -> species_raw:", species_raw.shape, "| obs_raw:", obs_raw.shape)
```

âœ… Both datasets were loaded successfully:
- `species_info.csv` has ~5800 rows
- `observations.csv` has ~23,000 rows  

---

## Step 0.2 â€” Standardize columns & strip text

We clean column names, strip whitespace, and ensure that observation counts are numeric.


```python
# ======================
# STEP 0.2 â€” Standardize columns & strip text (without breaking NaN)
# ======================
species = species_raw.rename(columns={
    "category": "category",
    "scientific_name": "scientific_name",
    "common_names": "common_names",
    "conservation_status": "conservation_status"
}).copy()

obs = obs_raw.rename(columns={
    "scientific_name": "scientific_name",
    "park_name": "park_name",
    "observations": "observations"
}).copy()

# Strip whitespace on text columns â€” do NOT cast to str (preserves NaN)
for col in ["category", "scientific_name", "common_names", "conservation_status"]:
    if col in species.columns and species[col].dtype == object:
        species[col] = species[col].str.strip()

for col in ["scientific_name", "park_name"]:
    if col in obs.columns and obs[col].dtype == object:
        obs[col] = obs[col].str.strip()

# Ensure observations is numeric
obs["observations"] = pd.to_numeric(obs["observations"], errors="coerce").fillna(0).astype(int)

# If the source had literal "nan" strings that slipped through, normalize them to real NaN
species["conservation_status"] = species["conservation_status"].replace(
    {"nan": np.nan, "NaN": np.nan, "": np.nan}
)
```

âœ… The data is now tidy:
- Observations are integers
- Missing conservation statuses are real `NaN` values
- Whitespace has been removed from text fields

---


## Step 0.3 â€” Create clean conservation status & protection flag

We replace missing values with `"No Intervention"` and create a flag `is_protected`.


```python
# ======================
# STEP 0.3 â€” Clean status & create flags
#   - Replace missing conservation_status with "No Intervention"
#   - Flag is_protected = True if status != "No Intervention"
# ======================
species["status_clean"] = species["conservation_status"].fillna("No Intervention")
species["is_protected"] = species["status_clean"].ne("No Intervention")

# Verify we now have both True/False
print("\nstatus_clean unique values:", species["status_clean"].unique())
print("is_protected counts:\n", species["is_protected"].value_counts(dropna=False))
```

âœ… Most species have `"No Intervention"`.  
Only a small fraction (~200) are marked as **protected**.  

---


## Step 0.4 â€” Deduplicate species

Some species may appear multiple times. We keep the first record per `(scientific_name, category)`.


```python
# ======================
# STEP 0.4 â€” Deduplicate species to prevent exploding joins
#   - Keep first record per (scientific_name, category)
# ======================
species = (
    species.sort_values(["scientific_name", "category"])
           .drop_duplicates(subset=["scientific_name", "category"], keep="first")
)
```

âœ… Most species have `"No Intervention"`.  
Only a small fraction (~200) are marked as **protected**.  

--

## Step 0.5 â€” Merge observations with species metadata

We join the species metadata with observations to get a unified dataset.


```python
# ======================
# STEP 0.5 â€” Merge observations with species metadata
# ======================
df = obs.merge(
    species[["scientific_name", "category", "status_clean", "is_protected"]],
    on="scientific_name",
    how="left"
)
```

---


## Step 0.6 â€” Sanity checks


```python
# ======================
# STEP 0.6 â€” Sanity checks
# ======================
summary = {
    "species_raw_shape": species_raw.shape,
    "obs_raw_shape": obs_raw.shape,
    "species_shape_clean": species.shape,
    "obs_shape_clean": obs.shape,
    "merged_shape": df.shape,
    "unique_species_in_obs": df["scientific_name"].nunique(),
    "unique_parks": df["park_name"].nunique(),
    "total_observations": int(df["observations"].sum()),
    "missing_after_merge": df.isna().sum().to_dict(),
}
print("\n=== STEP 0 SUMMARY ===")
for k, v in summary.items():
    print(f"{k}: {v}")

print("\nSpecies (clean) â€” head():")
print(species.head(3))
print("\nObservations (clean) â€” head():")
print(obs.head(3))
print("\nMerged df (obs + species) â€” head():")
print(df.head(3))
```

âœ… The cleaned dataset:
- ~5500 unique species
- 4 national parks
- ~3.3 million total observations
- No missing values after merging

---


```python
# ======================
# Q1 â€” Protected vs. not protected (overall)
# ======================
protected_summary = (
    species.groupby("is_protected")
           .size()
           .reset_index(name="count")
)

print("\nQ1 â€” Protected vs. not protected (overall):")
print(protected_summary)
```

```python
# ======================
# Q2 â€” Protected vs Not Protected by Category + KPI
# ======================

# 1) Count species by (category, is_protected)
category_summary = (
    species.groupby(["category", "is_protected"])
           .size()
           .reset_index(name="count")
)

# 2) Pivot so we get columns for True/False
cat_pivot = (
    category_summary
    .pivot(index="category", columns="is_protected", values="count")
    .fillna(0)
    .astype(int)
    .rename(columns={False: "not_protected", True: "protected"})
)

# 3) KPI: protection ratio per category
cat_pivot["protection_ratio"] = (
    cat_pivot["protected"] / (cat_pivot["protected"] + cat_pivot["not_protected"]).replace({0: np.nan})
)

# 4) Sort by highest ratio
cat_pivot = cat_pivot.sort_values("protection_ratio", ascending=False)

print("Q2 â€” Protected vs Not Protected by Category (with protection_ratio):")
print(cat_pivot)
```

```python
# ======================
# Q3 â€” Protected Species per Park
# ======================

protected_by_park = (
    df[df["is_protected"]]
    .groupby("park_name")["scientific_name"]
    .nunique()   # unique species, not just rows
    .reset_index(name="protected_species_count")
    .sort_values("protected_species_count", ascending=False)
)

print("Q3 â€” Protected species count per park:")
print(protected_by_park)
```

```python
sns.set(style="whitegrid")
ax = sns.barplot(
    data=protected_by_park,
    x="park_name",
    y="protected_species_count",
    palette="viridis"
)
ax.set_title("Protected Species Count per Park")
ax.set_xlabel("National Park")
ax.set_ylabel("Protected Species Count")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```

```python
# ======================
# Q4 â€” Protection rate by category
# ======================

# Count how many species are protected vs. not, grouped by category
category_protection = (
    species.groupby(["category", "is_protected"])
           .size()
           .reset_index(name="count")
)

# Pivot for easier calculation
category_pivot = category_protection.pivot(
    index="category",
    columns="is_protected",
    values="count"
).fillna(0)

# Rename columns
category_pivot = category_pivot.rename(columns={False: "not_protected", True: "protected"})

# Calculate protection rate = protected / total
category_pivot["protection_rate"] = (
    category_pivot["protected"] / (category_pivot["protected"] + category_pivot["not_protected"])
)

print("\nQ4 â€” Protection rate by category:")
print(category_pivot.sort_values("protection_rate", ascending=False))

# Optional: visualize
plt.figure(figsize=(8,6))
sns.barplot(
    data=category_pivot.reset_index(),
    x="category", y="protection_rate",
    palette="viridis"
)
plt.title("Protection Rate by Species Category")
plt.ylabel("Proportion Protected")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

```python
# ======================
# Q5 â€” Top observed species overall
# ======================

# Aggregate observations across all parks
species_obs = (
    df.groupby(["scientific_name", "category", "status_clean"])
      .agg(total_obs=("observations", "sum"))
      .reset_index()
)

# Sort descending to get the most observed species
top_species = species_obs.sort_values("total_obs", ascending=False).head(10)

print("\nQ5 â€” Top 10 observed species overall:")
print(top_species)

# Optional: visualize top 10 species
plt.figure(figsize=(10,6))
sns.barplot(
    data=top_species,
    x="total_obs", y="scientific_name",
    hue="category", dodge=False, palette="tab10"
)
plt.title("Top 10 Most Observed Species (All Parks)")
plt.xlabel("Total Observations")
plt.ylabel("Species")
plt.tight_layout()
plt.show()
```
